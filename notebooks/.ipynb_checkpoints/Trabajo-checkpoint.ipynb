```python
import ipywidgets as widgets 
import pandas as pd
import sys
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from pandas.tools.plotting import radviz
import pandas.io.sql as pd_sql
import sqlite3 as sql


%matplotlib inline
```

```python
data1 = pd.read_csv('data/proyectobi_Televisores_items.csv')
```

```python
data2 = pd.read_csv('data/proyectobi_Computadoras_items.csv')
```

```python
data3 = pd.read_csv('data/proyectobi_Licuadoras_items.csv')
```

```python
data4 = pd.read_csv('data/proyectobi_Impresoras_items.csv')
```

```python
data5 = pd.read_csv('data/proyectobi_Relojes_items.csv')
```

```python
frames = [data1, data2, data3, data4, data5] #Concatenar datas
```

```python
df = pd.concat(frames) #Concatenar datas
```

```python
from IPython.display import display

w = widgets.Dropdown(
    options=[u'Televisión y Video', u'Relojes', u'Computadoras y Servidores', u'Electrodomésticos', u'Impresoras y Accesorios'],
    value = u'Relojes',
    description='Productos:',
    disabled=False,
)
display(w)

#Widget de los productos 
```

```python
df[df[u'tipo']== w.value]
```

```python
ax = df.groupby(['tiempo']).size().sort_values(ascending=False)[:20].plot.bar(title = u'Tiempo de Venta', legend=True)

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))   
    
    #Grafica en base al atributo tiempo_venta, que muestra la cantidad de vendedores por el tiempo que van vendiendo
```

```python
condicion =['Nuevo', 'Usado'] #- Obtener un dataset que permita obtener todos los precios menores a 100 y que pertenezcan a la condicion de nuevo o usado
df[df['precio']< 100].query('condicion == @condicion')
```

```python
df.ubicacion.value_counts()[:5].sort_values(inplace=False).plot('barh'); #Obtener un gráfico que me permita vizualizar en base al atributo ubicacion
plt.title(u'Ubicación Vendedeores%')
```

```python
media = df["precio"].mean() #En base al atributo precio obtener el precio promedio de todos los productos
media
```

```python
plt.figure(figsize=(40,20))
# plt.title(u"Título de la gráfica", fontsize = 30)
plt.subplot2grid((2,3), (0,0))
df.reputacion.value_counts().plot(kind='bar')
plt.title(u'Recomendaciones por Reputación')
```

```python
plt.figure()
plt.title(u"Recomendaciones", fontsize = 30)
df.reputacion[df['tiempo'] == '5 Años'].value_counts().plot(kind='bar')
plt.show()

#Tipo de recomendaciones a vendedores que llevan menos de 5 años 
```
